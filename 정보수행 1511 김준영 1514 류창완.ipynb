```python
import csv
f = open('정보 data.csv')

data = csv.reader(f)
next(data)
next(data)
next(data)
data = list(data)

name = input('교통사고 비율을 알고 싶은 요일을 입력해주세요:')
pivot = []

for row in data:
    if name in row[0]:
        for i in range(3,len(row)):
            pivot.append(int(i))
            import matplotlib.pyplot as plt

            plt.figure(figsize=(10,8), dpi=300)
            plt.rc('font', family='Malgun Gothic')
            size = [row[3],row[4],row[5],row[6],row[7],row[8],row[9],row[10],row[11],row[12],row[13],row[14]]
            labels = '0시~2시', '2시~4시','4시~6시','6시~8시','8시~10시','10시~12시','12시~14시','14시~16시','16시~18시','18시~20시','20시~22시','22시~24시'
        
            plt.title('교통사고 비율')
            plt.pie(size, labels=labels, autopct = '%1.1f%%', startangle=90)
        
            plt.axis('equal')
            plt.show()
```
